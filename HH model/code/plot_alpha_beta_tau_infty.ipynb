# Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Define Functions

```python
# Define the functions for alpha and beta
def alpha_n(V):
    return 0.01 * (V + 10) / (np.exp(0.1 * (V + 10)) - 1)

def beta_n(V):
    return 0.125 * np.exp(V / 80)

def alpha_m(V):
    return 0.1 * (V + 25) / (np.exp(0.1 * (V + 25)) - 1)

def beta_m(V):
    return 4 * np.exp(V / 18)

def alpha_h(V):
    return 0.07 * np.exp(V / 20)

def beta_h(V):
    return 1 / (1 + np.exp(0.1 * (V + 30)))
```

```python
# Define tau and infty functions
def tau_n(V):
    return 1 / (alpha_n(V) + beta_n(V))

def infty_n(V):
    return alpha_n(V) / (alpha_n(V) + beta_n(V))

def tau_m(V):
    return 1 / (alpha_m(V) + beta_m(V))

def infty_m(V):
    return alpha_m(V) / (alpha_m(V) + beta_m(V))

def tau_h(V):
    return 1 / (alpha_h(V) + beta_h(V))

def infty_h(V):
    return alpha_h(V) / (alpha_h(V) + beta_h(V))

# Plot
def plot_alpha_and_beta(V):
    plt.figure(figsize=(15, 8))

    plt.subplot(2, 3, 1)
    plt.plot(V, alpha_n(V), label="α_n")
    plt.legend()
    plt.title('alpha_n')

    plt.subplot(2, 3, 2)
    plt.plot(V, alpha_m(V), label="α_m")
    plt.legend()
    plt.title('alpha_m')

    plt.subplot(2, 3, 3)
    plt.plot(V, alpha_h(V), label="α_h")
    plt.legend()
    plt.title('alpha_h')

    plt.subplot(2, 3, 4)
    plt.plot(V, beta_n(V), color='orange', label="β_n")
    plt.legend()
    plt.title('beta_n')

    plt.subplot(2, 3, 5)
    plt.plot(V, beta_m(V), color='orange', label="β_m")
    plt.legend()
    plt.title('beta_m')

    plt.subplot(2, 3, 6)
    plt.plot(V, beta_h(V), color='orange', label="β_h")
    plt.legend()
    plt.title('beta_h')

    plt.show()

# 
def plot_tau_and_infty(V):
    plt.figure(figsize=(15, 8))

    plt.subplot(2, 3, 1)
    plt.plot(V, tau_n(V), label="τ_n")
    plt.legend()
    plt.title('τ_n vs V')

    plt.subplot(2, 3, 2)
    plt.plot(V, tau_m(V), label="τ_m")
    plt.legend()
    plt.title('τ_m vs V')

    plt.subplot(2, 3, 3)
    plt.plot(V, tau_h(V), label="τ_h")
    plt.legend()
    plt.title('τ_h vs V')

    plt.subplot(2, 3, 4)
    plt.plot(V, infty_n(V), label="n(∞)", color='orange')
    plt.legend()
    plt.title('n(∞) vs V')

    plt.subplot(2, 3, 5)
    plt.plot(V, infty_m(V), label="m(∞)", color='orange')
    plt.legend()
    plt.title('m(∞) vs V')

    plt.subplot(2, 3, 6)
    plt.plot(V, infty_h(V), label="h(∞)", color='orange')
    plt.legend()
    plt.title('h(∞)vs V')

    plt.show()
```

# Plot

```python
# Generate voltage values
V = np.linspace(-100, 100, 400)

# Plot alpha and beta
plot_alpha_and_beta(V)

# Plot τ and ∞
plot_tau_and_infty(V)
```
